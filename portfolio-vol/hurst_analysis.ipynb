Estimate the Hurst exponent H for each ticker via classic R/S analysis. H is the slope of log₂(R/S) regressed on log₂(T):

- **H = 0.5** → independent / random walk
- **H > 0.5** → trending (R/S grows faster than √T)
- **H < 0.5** → mean-reverting (R/S grows slower than √T)

Trader's punchline: H ≠ 0.5 means √T vol annualization is wrong for that asset.

Edit the `TICKERS` / `START` / `END` in the second code cell, then **Run All**. Works in Google Colab (one-click) and locally in VS Code / Jupyter.

```python
from hurst import run_hurst_analysis

TICKERS = ["SPY", "QQQ", "TLT", "GLD"]

START = "2020-01-01"
END   = "2025-01-01"

ROLLING_WINDOW = 252   # trailing days used for the rolling-H chart
T_VALUES = None        # None = auto-pick based on sample size (powers of 2)

result = run_hurst_analysis(
    TICKERS,
    start=START,
    end=END,
    rolling_window=ROLLING_WINDOW,
    T_values=T_VALUES,
)
```

`result` holds everything for further poking:
- `result.per_ticker["SPY"].H` — Hurst exponent
- `result.per_ticker["SPY"].fit.rs_table` — the full R/S table at each T
- `result.rolling` — DataFrame of rolling H per ticker (date × ticker)
- `result.returns` — the underlying log-returns panel
